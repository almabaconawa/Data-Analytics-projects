**1. You have a population consisting of three numbers: 5, 10 and 15. Two numbers are randomly selected with replacement from this population. List all possible pairs that can be selected and calculate the sampling distribution of the sample mean. Then, plot a histogram to visualize the distribution of the sample means.**

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt

# Population
population = [5, 10, 15]

# Generate all possible pairs with replacement
pairs = list(itertools.product(population, repeat=2))
print("All possible pairs:")
for pair in pairs:
    print(pair)

# Calculate the sample mean for each pair
sample_means = [np.mean(pair) for pair in pairs]
print("\nSample means:")
print(sample_means)

# Calculate the sampling distribution
unique_means, counts = np.unique(sample_means, return_counts=True)
probabilities = counts / len(sample_means)

print("\nSampling Distribution of the Sample Mean:")
for mean, prob in zip(unique_means, probabilities):
    print(f"Mean: {mean}, Probability: {prob}")

# Plot histogram of the sampling distribution
plt.figure(figsize=(8,5))
plt.bar(unique_means, probabilities, width=1, color='skyblue', edgecolor='black')
plt.xlabel("Sample Mean")
plt.ylabel("Probability")
plt.title("Sampling Distribution of the Sample Mean")
plt.xticks(unique_means)
plt.grid(axis='y')
plt.show()

# itertools.product(population, repeat=2) generates all possible pairs with replacement.
# np.mean(pair) calculates the mean of each pair.
# np.unique(..., return_counts=True) computes probabilities of each unique sample mean.
# The histogram visualizes the sampling distribution of the sample mean.
```

**2.  Generate 1000 random samples from a uniform distribution between 0 and 1. Compute the mean of each sample with a sample size of 30. Plot the histogram of these means and overlay a normal distribution curve. Note: In this question, we want to generate 1000 random samples from a uniform distribution (which is not normally distributed) and calculate the mean of each sample. According to the Central Limit Theorem, the distribution of these sample means will approximate a normal distribution as the number of samples increases, regardless of the original distribution's shape. By plotting the histogram of these sample means and overlaying a normal distribution curve, we can observe this approximation and see the Central Limit Theorem in action.**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
n_samples = 1000   # Number of random samples
sample_size = 30   # Sample size for each mean

# Store sample means
sample_means = []

# Generate 1000 sample means
for _ in range(n_samples):
    sample = np.random.uniform(0, 1, sample_size)  # Uniform distribution between 0 and 1
    sample_means.append(np.mean(sample))

# Convert to numpy array for convenience
sample_means = np.array(sample_means)

# Plot histogram of sample means
plt.figure(figsize=(10,6))
plt.hist(sample_means, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7, label='Sample Means')

# Overlay normal distribution curve
mu = np.mean(sample_means)           # Mean of the sample means
sigma = np.std(sample_means)         # Standard deviation of the sample means
x = np.linspace(min(sample_means), max(sample_means), 1000)
plt.plot(x, norm.pdf(x, mu, sigma), color='red', linewidth=2, label='Normal Approximation')

# Labels and title
plt.title("Central Limit Theorem: Sample Means from Uniform Distribution")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()

# np.random.uniform(0,1,sample_size) generates a uniform random sample of size 30.
# We repeat this 1000 times to get 1000 sample means.
# According to the Central Limit Theorem, the distribution of these means
# will approximate a normal distribution, even though the original data is uniform.
# Overlaying norm.pdf() shows the normal approximation visually.
```

**3. Repeat the above experiment, but this time use sample sizes of 10, 50, and 100. Plot the histograms of the sample means for each sample size on separate plots. Note: In this question we want to extend the first one by varying the sample size (10, 50, and 100). The Central Limit Theorem suggests that as the sample size increases, the distribution of the sample means will become closer to a normal distribution. By plotting histograms for different sample sizes, students can visualize how increasing the sample size affects the normality of the distribution of sample means. Smaller sample sizes may show more deviation from normality compared to larger sample sizes, demonstrating the Central Limit Theorem’s reliance on larger samples for a normal approximation.**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
n_samples = 1000           # Number of random samples
sample_sizes = [10, 50, 100]  # Different sample sizes

# Repeat the experiment for each sample size
for sample_size in sample_sizes:
    sample_means = []

    # Generate sample means
    for _ in range(n_samples):
        sample = np.random.uniform(0, 1, sample_size)
        sample_means.append(np.mean(sample))

    sample_means = np.array(sample_means)

    # Plot histogram of sample means
    plt.figure(figsize=(8,5))
    plt.hist(sample_means, bins=30, density=True, color='skyblue', edgecolor='black', alpha=0.7, label='Sample Means')

    # Overlay normal distribution curve
    mu = np.mean(sample_means)
    sigma = np.std(sample_means)
    x = np.linspace(min(sample_means), max(sample_means), 1000)
    plt.plot(x, norm.pdf(x, mu, sigma), color='red', linewidth=2, label='Normal Approximation')

    # Labels, title, legend
    plt.title(f"CLT: Sample Means (Sample Size = {sample_size})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

# We generate 1000 sample means for each sample size (10, 50, 100) from a uniform distribution.
# Histograms show the distribution of these sample means.
# According to the Central Limit Theorem:
# Smaller samples (n=10) may show deviation from normality.
# Larger samples (n=50, 100) produce a distribution closer to normal.
# Overlaying norm.pdf() illustrates the normal approximation for each sample size.
```

**4. Generate 1000 random samples from a binomial distribution with 20 trials and a success probability of 0.3. Compute the mean of each sample with a sample size of 50. Plot the histogram of these means and overlay a normal distribution curve. Note: In this question, we use a binomial distribution, which can be skewed depending on the probability of success and the number of trials. By computing the means of samples from this non-normal distribution, we illustrate that even if the original data is not normally distributed, the distribution of sample means will approach normality if the sample size is sufficiently large. This demonstrates the Central Limit Theorem's key point that the sample means will tend toward a normal distribution regardless of the original data distribution.**


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
n_samples = 1000       # Number of sample means
sample_size = 50       # Sample size for each mean
n_trials = 20          # Number of trials in binomial
p_success = 0.3        # Probability of success in each trial

# Store sample means
sample_means = []

# Generate 1000 sample means
for _ in range(n_samples):
    sample = np.random.binomial(n_trials, p_success, sample_size)
    sample_means.append(np.mean(sample))

sample_means = np.array(sample_means)

# Plot histogram of sample means
plt.figure(figsize=(10,6))
plt.hist(sample_means, bins=30, density=True, color='lightgreen', edgecolor='black', alpha=0.7, label='Sample Means')

# Overlay normal distribution curve
mu = np.mean(sample_means)
sigma = np.std(sample_means)
x = np.linspace(min(sample_means), max(sample_means), 1000)
plt.plot(x, norm.pdf(x, mu, sigma), color='red', linewidth=2, label='Normal Approximation')

# Labels and title
plt.title("Central Limit Theorem: Sample Means from Binomial Distribution")
plt.xlabel("Sample Mean")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()


# np.random.binomial(n_trials, p_success, sample_size) generates random samples from a binomial distribution.
# We repeat this 1000 times to get 1000 sample means.
# The histogram visualizes the distribution of sample means.
# According to the Central Limit Theorem, the distribution of these means will approximate a normal distribution,
# even though the original binomial distribution may be skewed.
# Overlaying norm.pdf() demonstrates the normal approximation visually.
```

**5. Load the mtcars dataset. Compute the means of the mpg (miles per gallon) variable for 1000 random samples each with a size of 10. Plot the histogram of these means and overlay a normal distribution curve. Additionaly, try repeating the analysis with different sample sizes (e.g., 5, 20, 50) and observe how the distribution of sample means changes with varying sample sizes. Note: In this question, we use a real dataset (mtcars) to compute the means of the mpg variable. By analyzing the sample means of a real-world dataset, we provide a practical demonstration of the Central Limit Theorem. This exercise shows that even with real, empirical data, the distribution of sample means tends to be normal, thus illustrating the theorem's applicability in real-world scenarios.**




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns

# Load the mtcars dataset from seaborn
# Note: seaborn does not have mtcars, so we can use it from statsmodels
import statsmodels.api as sm
mtcars = sm.datasets.get_rdataset("mtcars").data

# Extract the mpg column
mpg = mtcars['mpg'].values

# Parameters
n_samples = 1000        # Number of sample means
sample_sizes = [5, 10, 20, 50]  # Different sample sizes to try

# Repeat analysis for each sample size
for sample_size in sample_sizes:
    sample_means = []

    # Generate 1000 sample means
    for _ in range(n_samples):
        sample = np.random.choice(mpg, size=sample_size, replace=True)
        sample_means.append(np.mean(sample))

    sample_means = np.array(sample_means)

    # Plot histogram of sample means
    plt.figure(figsize=(8,5))
    sns.histplot(sample_means, bins=30, kde=False, stat='density', color='lightblue', edgecolor='black', alpha=0.7, label='Sample Means')

    # Overlay normal distribution curve
    mu = np.mean(sample_means)
    sigma = np.std(sample_means)
    x = np.linspace(min(sample_means), max(sample_means), 1000)
    plt.plot(x, norm.pdf(x, mu, sigma), color='red', linewidth=2, label='Normal Approximation')

    # Labels and title
    plt.title(f"CLT: Sample Means of MPG (Sample Size = {sample_size})")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()


# np.random.choice() samples from the 'mpg' column with replacement to compute sample means.
# Repeating 1000 times gives a distribution of sample means for each sample size.
# Histograms show how the distribution of sample means changes as the sample size increases.
# According to the Central Limit Theorem:
#   * Smaller sample sizes (e.g., 5) may show a less smooth, slightly skewed distribution.
#   * Larger sample sizes (20, 50) produce distributions that approximate a normal distribution more closely.
# Overlaying norm.pdf() helps visualize this normal approximation.
```
