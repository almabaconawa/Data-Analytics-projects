**1. Generate 1000 random values from a normal distribution with a mean of 5 and a standard deviation of 2. Calculate the sample mean and standard deviation. Compare them with the theoretical values.**

```python
import numpy as np
mu = 5
sigma = 2
size = 1000
normal_data = np.random.normal(mu, sigma, size)
sample_mean = normal_data.mean()
sample_std = normal_data.std()
theoretical_mean = mu
theoretical_std = sigma

print("Sample mean:", round(sample_mean, 2))
print("Sample standard deviation:", round(sample_std, 2))
print("Theoretical mean:", theoretical_mean)
print("Theoretical standard deviation:", theoretical_std)
```

```python
import numpy as np

# Parameters
mu = 5        # mean
sigma = 2     # standard deviation
size = 1000   # number of random values

# Generate random values from Normal distribution
normal_data = np.random.normal(mu, sigma, size)

# Calculate sample mean and std
sample_mean = normal_data.mean()
sample_std = normal_data.std()

# Theoretical values
theoretical_mean = mu
theoretical_std = sigma

print("Sample mean:", round(sample_mean, 2))
print("Sample standard deviation:", round(sample_std, 2))
print("Theoretical mean:", theoretical_mean)
print("Theoretical standard deviation:", theoretical_std)

# np.random.normal(mu, sigma, size) generates random values
# from a normal distribution with given mean and standard deviation.
# The sample mean and standard deviation should be close to the theoretical
# values (5 and 2), especially with large sample size.
```

**3. Suppose the normal distribution has a mean of 5 and a standard deviation of 2. What is the probability of drawing a value less than 3 from this distribution? Use the cumulative distribution function (CDF) in Python to calculate the probability.**


```python
from scipy.stats import norm

mu = 5
sigma = 2

# Cumulative Distribution Function at x = 3
prob = norm.cdf(3, loc=mu, scale=sigma)

print("Probability of drawing a value less than 3:", prob)


# norm.cdf(x, loc=mu, scale=sigma) gives P(X ≤ x).
# # When we describe a normal distribution, the standard way is: N(μ, σ^2)
# μ = mean (center of the distribution)
# σ^2 = variance (square of the standard deviation)
# Here it calculates P(X < 3) for N(5, 2^2).
```

**4.  For a normal distribution with a mean of 5 and a standard deviation of 2, compute the 10th, 25th, 50th (median), 75th, and 90th percentiles. Use Python’s norm.ppf() function to calculate these quantiles and display their values.**


```python
from scipy.stats import norm

# Parameters
mu = 5      # mean
sigma = 2   # standard deviation

# Percentiles we want
percentiles = [0.10, 0.25, 0.50, 0.75, 0.90]

# Compute quantiles using ppf (inverse CDF)
quantile_values = [norm.ppf(p, loc=mu, scale=sigma) for p in percentiles]

# Display results
for p, q in zip(percentiles, quantile_values):
    print(f"{int(p*100)}th percentile: {q:.2f}")

# norm.ppf(p, loc=μ, scale=σ) gives the value x such that P(X ≤ x) = p.
# The 50th percentile (p=0.5) is the median of the distribution.
```

**5. Using the same normal distribution with a mean of 5 and a standard deviation of 2, plot the normal distribution and mark the 10th, 25th, 50th (median), 75th, and 90th percentiles on the graph. Use vertical lines to denote these percentiles and include labels.**




```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 5      # mean
sigma = 2   # standard deviation

# Create x values for the curve
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, loc=mu, scale=sigma)

# Percentiles we want
percentiles = [0.10, 0.25, 0.50, 0.75, 0.90]
quantile_values = [norm.ppf(p, loc=mu, scale=sigma) for p in percentiles]

# Plot the normal distribution curve
plt.plot(x, y, label="Normal PDF", color="blue")

# Add vertical lines for percentiles
for p, q in zip(percentiles, quantile_values):
    plt.axvline(q, color="red", linestyle="--", linewidth=1)
    plt.text(q, norm.pdf(q, mu, sigma) + 0.005, f"{int(p*100)}%",
             ha="center", color="red")

# Labels and title
plt.title("Normal Distribution with Percentiles")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()

# norm.pdf(x, μ, σ) plots the probability density function.
# norm.ppf(p, μ, σ) finds the quantile (value at percentile p).
# Vertical dashed lines show where each percentile lies.
```

**6. For a normal distribution with a mean of 0 and a standard deviation of 1, plot the normal distribution and shade the area between -2 and 2. Add vertical lines at -2 and 2 to show the boundaries of this shaded region. Include a legend to indicate the shaded area.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for standard normal distribution
mu = 0
sigma = 1

# Create x values for the curve
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)
y = norm.pdf(x, loc=mu, scale=sigma)

# Plot the standard normal curve
plt.plot(x, y, color="blue", label="Standard Normal PDF")

# Shade area between -2 and 2
x_shade = np.linspace(-2, 2, 1000)
y_shade = norm.pdf(x_shade, mu, sigma)
plt.fill_between(x_shade, y_shade, color="orange", alpha=0.4, label="Area between -2 and 2")

# Add vertical lines at -2 and 2
plt.axvline(-2, color="red", linestyle="--", label="-2")
plt.axvline(2, color="green", linestyle="--", label="2")

# Add labels, title, legend, and grid
plt.title("Standard Normal Distribution with Shaded Area (-2 to 2)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()

# norm.pdf(x, μ, σ) plots the probability density function.
# fill_between() shades the area under the curve for x values between -2 and 2.
# axvline() adds vertical lines to mark the boundaries of the shaded region.
# This shaded area visually represents P(-2 ≤ X ≤ 2) for a standard normal distribution.
```

**7. For a standard normal distribution with mean = 0 and standard deviation = 1, calculate the following probabilities:
a) The probability of drawing a value between -1 and 1.
b) The probability of drawing a value between -2 and 2.**

```python
from scipy.stats import norm

# Standard normal parameters
mu = 0
sigma = 1

# a) Probability between -1 and 1
prob_a = norm.cdf(1, loc=mu, scale=sigma) - norm.cdf(-1, loc=mu, scale=sigma)
print("P(-1 ≤ X ≤ 1):", round(prob_a, 4))

# b) Probability between -2 and 2
prob_b = norm.cdf(2, loc=mu, scale=sigma) - norm.cdf(-2, loc=mu, scale=sigma)
print("P(-2 ≤ X ≤ 2):", round(prob_b, 4))

# norm.cdf(x, loc=μ, scale=σ) calculates P(X ≤ x).
# To find P(a ≤ X ≤ b), subtract: P(X ≤ b) - P(X ≤ a).
# These probabilities show the proportion of values within 1 and 2 standard deviations
# from the mean in a standard normal distribution.
```

**8. For a standard normal distribution with mean = 0 and sd = 1, and another normal distribution with mean = 3 and sd = 2, plot both distributions on the same graph. Use different colors and line styles, and include a legend to distinguish them.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for both distributions
mu1, sigma1 = 0, 1   # Standard normal
mu2, sigma2 = 3, 2   # Second normal distribution

# Create x values covering both distributions
x = np.linspace(-5, 10, 1000)

# Compute PDFs
y1 = norm.pdf(x, loc=mu1, scale=sigma1)
y2 = norm.pdf(x, loc=mu2, scale=sigma2)

# Plot both distributions
plt.plot(x, y1, color='blue', linestyle='-', label='N(0,1)')
plt.plot(x, y2, color='red', linestyle='--', label='N(3,2^2)')

# Labels, title, legend, grid
plt.title("Comparison of Two Normal Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()

# norm.pdf(x, μ, σ) calculates the probability density function for each x.
# Different colors and line styles distinguish the two distributions.
# The legend helps identify which line corresponds to which distribution.
```
